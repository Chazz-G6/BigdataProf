```python
def plot_fff(fff, x, y, title="", xlabel='Date', ylabel='Close', dpi=100):
    plt.figure(figsize=(16,5), dpi=dpi)
    plt.plot(x, y, color='tab:red')
    plt.gca().set(title=title, xlabel=xlabel, ylabel=ylabel)
    plt.show()

    plot_fff(fff, x=fff.Date, y=fff.Close, title='Faang')
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
from pandas_datareader import data  
from datetime import datetime
import pandas as pd
import cufflinks
import plotly.offline as plyo

# 기본 csv 파일 불러오기
#file_path = "C:/Bigdata/work/Project01/faang_stocks_pandemic_data.csv" # 실습실
file_path = "C:/Users/Jiuk Choi/Documents/python/faang_stocks_pandemic_data.csv" # 집
ff = pd.read_csv(file_path,parse_dates=['Date'], index_col='Date')

```

```python
ff = ff.reset_index()

ff['Date'] = ff['Date'].apply(lambda x : datetime.strftime(x, '%Y-%m-%d')) # Datetime to str


stock_name = 'Facebook'
fig = go.Figure(data=[go.Candlestick(x=ff['Date'],
                                    open=ff['Open'],
                                    high=ff['High'],
                                    low=ff['Low'],
                                    close=ff['Close'])])

fig.layout = dict(title=stock_name, 
                       xaxis = dict(type="category", 
                                    categoryorder='category ascending'))
fig.update_xaxes(nticks=5)
```

```python
#-----------------------------------------------------------------------





%matplotlib inline

import plotly.offline as plyo
import cufflinks as cf
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import datetime
import mplfinance as mpf
# 기본 csv 파일 불러오기
file_path = "C:/Bigdata/work/Project01/faang_stocks_pandemic_data.csv" # 실습실
#file_path = "C:/Users/Jiuk Choi/Documents/python/faang_stocks_pandemic_data.csv" # 집
datas = pd.read_csv(file_path)
df = datas
df.Date = pd.to_datetime(df.Date)
df = df.set_index('Name')
```

```python
df_f = df.loc['Facebook']
df = df_f
```

```python
df = df.set_index('Date')
```

```python
mpf.plot(df)
```

```python
mpf.plot(df, type='line', volume=True)
```

```python
mpf.plot(df['2021-01':'2021-12'], figratio=(20,12),
        type='candle',
        title='Faang Price in 2021',
        mav=(20),volume=True,
        tight_layout=True,
        style='yahoo')
```

```python
plyo.iplot(df.iplot(asFigure=True))
```

```python

qf = cf.QuantFig(
        df,
        title='Facebook chart',
        legend='top',
        name='Facebook chart',
        up_color = 'red',    # 상승은 빨강색, 하락은 파랑색
        down_color = 'blue'
)
plyo.iplot(qf.iplot(asFigure=True))
```
