```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3], [4, 4, 4], '-', color='C0', label='Solid') #(solid) 
plt.plot([1, 2, 3], [3, 3, 3], '--', color='C0', label='Dashed')#Dash 
plt.plot([1, 2, 3], [2, 2, 2], ':', color='C0', label='Dotted') #Dot...
plt.plot([1, 2, 3], [1, 1, 1], '-.', color='C0', label='Dash-dot')# -.Dash & Dot
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis([0.8, 3.2, 0.5, 5.0])
plt.legend(loc='upper right', ncol=4) #location =  ncol 4
plt.show()
```
